## **Calcular Média Móvel**

```sql
/* Período 5, 10, 20, 50, 200 dias */
SELECT [Date], TicketName, [Close],
  AVG([Close]) -- Definir o cálculo
    OVER(PARTITION BY TicketName ORDER BY [Date]   -- Definir Over
    ROWS BETWEEN 4 PRECEDING AND CURRENT ROW) -- Definir Período
    AS dia5_media_movel, -- Definir Alias
  AVG([Close]) -- Definir o cálculo
    OVER(PARTITION BY TicketName ORDER BY [Date]   -- Definir Over
    ROWS BETWEEN 9 PRECEDING AND CURRENT ROW) -- Definir Período
    AS dia10_media_movel, -- Definir Alias
  AVG([Close]) -- Definir o cálculo
    OVER(PARTITION BY TicketName ORDER BY [Date]   -- Definir Over
    ROWS BETWEEN 19 PRECEDING AND CURRENT ROW) -- Definir Período
    AS dia20_media_movel, -- Definir Alias
  AVG([Close]) -- Definir o cálculo
    OVER(PARTITION BY TicketName ORDER BY [Date]   -- Definir Over
    ROWS BETWEEN 49 PRECEDING AND CURRENT ROW) -- Definir Período
    AS dia50_media_movel, -- Definir Alias
  AVG([Close]) -- Definir o cálculo
    OVER(PARTITION BY TicketName ORDER BY [Date]   -- Definir Over
    ROWS BETWEEN 199 PRECEDING AND CURRENT ROW) -- Definir Período
    AS dia200_media_movel
-- Definir Alias
FROM [SANDBOX].[dbo].[STOCK_HISTORY]
WHERE [Date] >= '2019'
```
